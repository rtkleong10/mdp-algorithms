```python
#for simulation

def sense(bot,bot_dir):
    global obstacles,map
    front = 2 #how far it can sense
    left = 5 #same as above
    right = 1 #same as above
    for i in range(front):
        if bot_dir=='E':
            x,y = bot[0]+2+i, bot[1]
        elif bot_dir=='N':
            x,y = bot[0], bot[1]+2+i
        elif bot_dir=='S':
            x,y = bot[0],bot[1]-2-i
        elif bot_dir=='W':
            x,y = bot[0]-2-i, bot[1]
        if x<0 or x>14 or y<0 or y>19:
            break
        if obstacles.get((x,y))!=None:
            map[y][x]=2
            break
        else:
            map[y][x]=1
    for i in range(front):
        if bot_dir=='E':
            x,y = bot[0]+2+i, bot[1]-1
        elif bot_dir=='N':
            x,y = bot[0]-1, bot[1]+2+i
        elif bot_dir=='S':
            x,y = bot[0]-1,bot[1]-2-i
        elif bot_dir=='W':
            x,y = bot[0]-2-i, bot[1]-1
        if x<0 or x>14 or y<0 or y>19:
            break
        if obstacles.get((x,y))!=None:
            map[y][x]=2
            break
        else:
            map[y][x]=1
    for i in range(front):
        if bot_dir=='E':
            x,y = bot[0]+2+i, bot[1]+1
        elif bot_dir=='N':
            x,y = bot[0]+1, bot[1]+2+i
        elif bot_dir=='S':
            x,y = bot[0]+1,bot[1]-2-i
        elif bot_dir=='W':
            x,y = bot[0]-2-i, bot[1]+1
        if x<0 or x>14 or y<0 or y>19:
            break
        if obstacles.get((x,y))!=None:
            map[y][x]=2
            break
        else:
            map[y][x]=1
        
    for i in range(left):
        if bot_dir=='S':
            x,y = bot[0]+2+i, bot[1]
        elif bot_dir=='E':
            x,y = bot[0], bot[1]+2+i
        elif bot_dir=='W':
            x,y = bot[0],bot[1]-2-i
        elif bot_dir=='N':
            x,y = bot[0]-2-i, bot[1]
        if x<0 or x>14 or y<0 or y>19:
            break
        if obstacles.get((x,y))!=None:
            map[y][x]=2
            break
        else:
            map[y][x]=1
            
    for i in range(right):
        if bot_dir=='N':
            x,y = bot[0]+2+i, bot[1]
        elif bot_dir=='W':
            x,y = bot[0], bot[1]+2+i
        elif bot_dir=='E':
            x,y = bot[0],bot[1]-2-i
        elif bot_dir=='S':
            x,y = bot[0]-2-i, bot[1]
        if x<0 or x>14 or y<0 or y>19:
            break
        if obstacles.get((x,y))!=None:
            map[y][x]=2
            break
        else:
            map[y][x]=1
            
```

```python
#find cord wrt the bot based on where it's facing
def findRightPos(bot_dir):
    if bot_dir=='E':
        return [0,-1]
    elif bot_dir=='N':
        return [1,0]
    elif bot_dir=='S':
        return [-1,0]
    elif bot_dir=='W':
        return [0,1]
                

#check for surroundings
def checkSurroundings(bot_dir,_dir):
    if bot_dir=='E':
        if _dir=='R':
            return [0,-2],[-1,-2],[1,-2]
        elif _dir=='L':
            return [0,2],[-1,2],[1,2]
        elif _dir=='F':
            return [2,0],[2,-1],[2,1]
    elif bot_dir=='N':
        if _dir=='R':
            return [2,0],[2,-1],[2,1]
        elif _dir=='L':
            return [-2,0],[-2,1],[-2,-1]
        elif _dir=='F':
            return [0,2],[-1,2],[1,2]
    elif bot_dir=='S':
        if _dir=='R':
            return [-2,0],[-2,1],[-2,-1]
        elif _dir=='L':
            return [2,0],[2,-1],[2,1]
        elif _dir=='F':
            return [0,-2],[-1,-2],[1,-2]
    elif bot_dir=='W':
        if _dir=='R':
            return [0,2],[-1,2],[1,2]
        elif _dir=='L':
            return [0,-2],[-1,-2],[1,-2]
        elif _dir=='F':
            return [-2,0],[-2,1],[-2,-1]
```

```python
#check if should turn right
def checkRight(bot,bot_dir,prev_pos):
    global map
    obstacle = False
    for x,y in checkSurroundings(bot_dir,'R'):
        if bot[1]+y<0 or bot[1]+y>19 or bot[0]+x<0 or bot[0]+x>14: #check wall
            return False
        elif map[bot[1]+y][bot[0]+x]==0: #check if right is unexplored
            return True
        elif map[bot[1]+y][bot[0]+x]==2: #check for obstacles
            obstacle= True
    if obstacle:
        return False
    add_x,add_y = findRightPos(bot_dir)
    rightpos = [bot[0]+add_x, bot[1]+add_y]
    if rightpos==prev_pos: 
        return False
    return True
    
    
#check if should move forward
def checkForward(bot,bot_dir):
    global map
    for x,y in checkSurroundings(bot_dir,'F'):
        if bot[1]+y<0 or bot[1]+y>19 or bot[0]+x<0 or bot[0]+x>14: #check wall
            return False
        elif map[bot[1]+y][bot[0]+x]==2: #check for obstacles
            return False
    return True

#check if should turn left
def checkLeft(bot,bot_dir):
    global map
    for x,y in checkSurroundings(bot_dir,'L'):
        if bot[1]+y<0 or bot[1]+y>19 or bot[0]+x<0 or bot[0]+x>14: #check wall
            return False
        elif map[bot[1]+y][bot[0]+x]==2: #check for obstacles
            return False
    return True
    
```

```python
#turn right
def turnRight(bot_dir):
    if bot_dir=='E':
        return 'S'
   
    elif bot_dir=='N':
        return 'E'
        
    elif bot_dir=='S':
        return 'W'
        
    elif bot_dir=='W':
        return 'N'

    
#move forward
def moveForward(bot,bot_dir):
    global map, prev_pos
    prev_pos = bot
    if bot_dir=='E':
        bot = [bot[0]+1, bot[1]]
    elif bot_dir=='N':
        bot = [bot[0],bot[1]+1]
    elif bot_dir=='S':
        bot = [bot[0],bot[1]-1]
    elif bot_dir=='W':
        bot = [bot[0]-1,bot[1]]
    return bot

#turn left
def turnLeft(bot_dir):
    if bot_dir=='E':
        return 'N'
   
    elif bot_dir=='N':
        return 'W'
        
    elif bot_dir=='S':
        return 'E'
        
    elif bot_dir=='W':
        return 'S'    
    
```

```python
def findUnexplored():
    global map
    for row in range(20):
        for col in range(15):
            if map[row][col]==0:
                return [col,row]
    return None
            
    
```

```python
def isPosSafe(x,y):
    global map
    if x<1 or x>13 or y<1 or y>18:
        return False
    for col in range(x-1,x+2):
        for row in range(y-1,y+2):
            if map[row][col]==2 or map[row][col]==0:
                return False
    return True
    
    
    
```

```python
def possiblePos(goal):
    d={}
    x,y = goal
    arr=[[0,-2],[-1,-2],[1,-2],[0,2],[-1,2],[1,2],[2,0],[2,-1],[2,1],[-2,0],[-2,1],[-2,-1]]
    for i in arr:
        X,Y = x+i[0],y+i[1]
        if isPosSafe(X,Y):
            d[(X,Y)]=1
    return d
```

```python
def fastestPath(bot,goal):
    import heapq
    global map
    x,y = bot
    cost = abs(goal[0]-x) + abs(goal[1]-y) 
    heap=[[cost,0,[[x,y]]]]
    visited = {}
    possiblePos={}
    possiblePos= possiblePos(goal)
    while heap:
        cost,move,path = heapq.heappop(heap)
        x,y = path[len(path)-1]
        if possiblePos.get(x,y)!=None:
            return path
        if len(path)>1:
            px,py = path[len(path)-2]
        else:
            px,py = x,y
        visited[(x,y)]=1
        if x>1:
            if visited.get((x-1,y))==None and map[y][x-1]!=2 and map[y][x-1]!=0:
                new_path= path+[[x-1,y]]
                turnCost = abs(x-1-px)%2 + abs(y-py)%2
                new_cost = abs(goal[0]-(x-1)) + abs(goal[1]-y) + move +1 + turnCost
                heapq.heappush(heap,[new_cost,move+1,new_path])
        if x<13:
            if visited.get((x+1,y))==None and map[y][x+1]!=2 and map[y][x+1]!=0:
                new_path= path+[[x+1,y]]
                turnCost = abs(x-1-px)%2 + abs(y-py)%2
                new_cost = abs(goal[0]-(x+1)) + abs(goal[1]-y) + move +1 + turnCost
                heapq.heappush(heap,[new_cost,move+1,new_path])
        if y>1:
            if visited.get((x,y-1))==None and map[y-1][x]!=2 and map[y-1][x]!=0:
                new_path= path+[[x,y-1]]
                turnCost = abs(x-1-px)%2 + abs(y-py)%2
                new_cost = abs(goal[0]-x) + abs(goal[1]-(y-1)) + move +1 + turnCost
                heapq.heappush(heap,[new_cost,move+1,new_path])
        if y<18:
            if visited.get((x,y+1))==None and map[y+1][x]!=2 and map[y+1][x]!=0:
                new_path= path+[[x,y+1]]
                turnCost = abs(x-1-px)%2 + abs(y-py)%2
                new_cost = abs(goal[0]-x) + abs(goal[1]-(y+1)) + move +1 + turnCost
                heapq.heappush(heap,[new_cost,move+1,new_path])
    
```
